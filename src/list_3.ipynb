# Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df["Diabetes_012"] = le.fit_transform(df["Diabetes_012"])
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=["Diabetes_012"])
y = df["Diabetes_012"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
```

# Model Training

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_score = lr_model.score(X_test, y_test)
lr_score
```

## K-Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
kn_model = KNeighborsClassifier()
```

```python
kn_model.fit(X_train, y_train)
```

```python
kn_score = kn_model.score(X_test, y_test)
kn_score
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(X_train, y_train)
```

```python
dt_score = dt_model.score(X_test, y_test)
dt_score
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
```

```python
rf_model.fit(X_train, y_train)
```

```python
rf_score = rf_model.score(X_test, y_test)
rf_score
```

```python
#sort the models by accuracy
models = [("Logistic Regression",lr_score),("K-Nearest Neighbours",kn_score),("Decision Tree",dt_score),("Random Forest",rf_score)]
models.sort(key=lambda x: x[1], reverse=True)
for i in models:
    print (i[0] + ":", i[1])
```
